```python
from bokeh.io import show, output_notebook
from bokeh.charts import Histogram
import numpy as np
import pandas as pd

output_notebook()

from_counts = np.load("Documents/from_counts.npy")
df = pd.DataFrame({'Votes':from_counts})
print(df.head())
hist = Histogram(from_counts, title="How Many Votes Made By Users", bins=12) 
show(hist)
```
